```python
import json
import matplotlib.pyplot as plt
```

```python
filename = "hance_clap-sanity_test-5K_soundly_2024_03_15-10_39_33"
with open(f"logs/{filename}.json", 'r') as fp:
    log = json.load(fp)

loss = []
eval = []
for epoch in log['epochs']:
    # Total loss divided by training steps
    # Training steps are determined by Pytorch
    loss.append(log['epochs'][epoch]['loss']) 
    eval.append(log['epochs'][epoch]['eval']) 

log['epochs']
```

```python
# Visually inspect the loss
plt.plot(loss)
plt.plot(eval)
```
