```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import pandas as pd
import datetime as dt
```

```python
ai_df = pd.read_csv("C:/Users/olive/Documents/Work/HANCE/Audio_ML_Industries_Guide/Infographics/ai-search_over_time.csv", header=1)
ai_dates = [dt.datetime.strptime(d,'%Y-%m-%d').date() for d in ai_df['Week']]
ai_searches = ai_df['AI: (Worldwide)']

ml_df = pd.read_csv("C:/Users/olive/Documents/Work/HANCE/Audio_ML_Industries_Guide/Infographics/ml-search_over_time.csv", header=1)
#print(ml_df)
ml_dates = [dt.datetime.strptime(d,'%Y-%m-%d').date() for d in ai_df['Week']]
ml_searches = ml_df['machine learning: (Worldwide)']

audio_ai_df = pd.read_csv("C:/Users/olive/Documents/Work/HANCE/Audio_ML_Industries_Guide/Infographics/audio_ai-search_over_time.csv", header=1)
audio_ai_dates = [dt.datetime.strptime(d,'%Y-%m-%d').date() for d in ai_df['Week']]
audio_ai_searches = audio_ai_df['audio ai: (Worldwide)']

enhance_audio_df = pd.read_csv("C:/Users/olive/Documents/Work/HANCE/Audio_ML_Industries_Guide/Infographics/enhance_audio-search_over_time.csv", header=1)
enhance_audio_dates = [dt.datetime.strptime(d,'%Y-%m-%d').date() for d in ai_df['Week']]
enhance_audio_searches = enhance_audio_df['enhance audio: (Worldwide)']
```

```python
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y'))
plt.gca().xaxis.set_major_locator(mdates.YearLocator())

frame_color = 'black'

plt.gca().set_title(
    "Queries Over Time", 
    fontdict={
        'fontsize': 16,
        'color':frame_color
    }
)

plt.locator_params(axis='x', nbins=1)
plt.grid(visible=True, axis='x')
plt.margins(x=0)

ax = plt.gca()
for spine in ax.spines:
    ax.spines[spine].set_color(frame_color)
    ax.tick_params(color='white', labelcolor=frame_color)

plt.xticks(rotation=90, ha='right')
plt.plot(ai_dates, ai_searches, label="'ai'", color='#004DC6')
plt.plot(ai_dates, audio_ai_searches, label="'audio ai'", color='#F87060')
plt.plot(ai_dates, ml_searches, label="'machine learning'", color='#B9E28C')
plt.plot(ai_dates, enhance_audio_searches, label="'enhance audio'", color='#43C59E')
plt.legend()
plt.gcf().autofmt_xdate()
plt.savefig('ai_trends_black.png',transparent=True, dpi=300)
```
